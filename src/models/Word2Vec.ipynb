```python
subjects = json.load(open('res/disciplina.json'))
pprint(subjects)
text = """
A palavra “deficiência” é acentuada pela mesma razão que
  a) sinônimo;
  b) excluído; 
  c) poética;
  d) privilégio;
  e) parênteses.
"""
```

```python
text_vector = phrase2vec(text)

distance_vector = []

for subject in subjects:
    subject_vector = phrase2vec(subject['value'].lower())
    distance_vector.append({
        'distance': la.norm(subject_vector - text_vector),
        'subject': subject['value']
    })
rank = sorted(distance_vector, key = lambda i: i['distance'])
print(rank[0]['subject'])
```
